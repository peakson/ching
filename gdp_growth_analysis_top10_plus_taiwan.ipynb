# GDP Growth Analysis - Top 10 Economies + Taiwan

這個 notebook 會：
1. 讀取 GDP 年增率資料（CSV）
2. 存入 pandas DataFrame
3. 用表格顯示資料
4. 畫出各國近五年的 GDP 年增率折線圖


```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 讀取 CSV 檔案
file_path = "gdp_growth_top10_plus_taiwan.csv"  # 確保這個檔案跟 notebook 放在同一個資料夾
df = pd.read_csv(file_path)

# 查看前幾列資料
df.head()
```

```python
# 以表格方式顯示全部資料
df
```

```python
# 將資料轉成：row = 年份, column = 國家, value = 成長率
gdp_pivot = df.pivot(index="Year", columns="Country", values="GDP_Growth")
gdp_pivot
```

```python
# 畫出各國近五年 GDP 年增率的折線圖
plt.figure(figsize=(10, 6))

gdp_pivot.plot(marker="o")

plt.title("GDP Growth Rate (Last 5 Years) - Top 10 Economies + Taiwan")
plt.xlabel("Year")
plt.ylabel("GDP Growth (%)")
plt.grid(True)
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
